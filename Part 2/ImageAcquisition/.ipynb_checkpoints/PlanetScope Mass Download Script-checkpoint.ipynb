```python
def sortPolygon(list):
    """
    Takes a list of x,y co-ordinates and converts it to a string of continuous points
    """
    string = ""
    for point in list:
        string += str(point[0]) + "," + str(point[1]) + ","
        
    return string[:-1]
```
